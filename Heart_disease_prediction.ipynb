```python
def outlier(df, columns=None):
    if columns == None:
        columns = df.columns
    for col in columns:
        mean = df[col].mean()
        std = df[col].std()
        
        border = np.abs(df[col] - mean) / std
        
        df = df[(border < 3.0)]
    
    return df
```

```python
def outlier_2(df, columns=None):
    if columns == None:
        columns = df.columns
    for col in columns:
        mean = df[col].mean()
        std = df[col].std()
        
        border = np.abs(df[col] - mean) / std
        
        df = df[(border < 1.9)]
    
    return df
```

```python
def outlier_3(df, columns=None):
    if columns == None:
        columns = df.columns
    for col in columns:
        mean = df[col].mean()
        std = df[col].std()
        
        border = np.abs(df[col] - mean) / std
        
        df = df[(border < 1.2)]
    
    return df
```

```python
#grid search
import time
t1 = time.time()

parameters = {  
    'n_estimators': [24,28,32],   
    'max_features': ('sqrt','log2','auto'), 
    'max_depth':    (5,10,15),     
    'criterion'   : ['gini', 'entropy'],
}
model = RandomForestClassifier()
gridsearch = GridSearchCV(estimator = model,     
                          param_grid = parameters, 
                          cv = 2,
                          n_jobs = -1,
                          return_train_score = False,
                          scoring = "accuracy"      
                         )
gridsearch.fit(X_train, y_train)

t2 = time.time()

print('Best params: {}'.format(gridsearch.best_params_)) 
print('Best Score: {}'.format(gridsearch.best_score_))
print((t2-t1)/60)
```
