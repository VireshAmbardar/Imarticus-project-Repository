```python
dd=df.copy()
```

```python
# using the price range as a factor for the sc_w replacements
```

```python
l=dd['price_range'].unique()
f=[0,0,0,0]
for i in l:
    a=dd[dd['price_range']==i]
    f[i]=a[a['sc_w']!=0]['sc_w'].median()
    print('PR :',i,' ',f[i])
```

```python
f   # median of sc_w for the price ranges 
```

```python
# Replacing the zeros from sc_w with resp. PR cate.
for i in [0,1,2,3]:
    dd['sc_w']=np.where(((dd['price_range']==i)&(dd['sc_w']==0)),int(f[i]),dd['sc_w'])
```

```python
dd[dd['sc_w']==0]
```

```python
dd.head()
```

```python
y=dd['price_range']
x=dd.drop(columns=['price_range'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=.3,random_state=88)
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(train_x,train_y)

print(knn.score(train_x,train_y))
print(knn.score(test_x,test_y))
```

```python
knn=KNeighborsClassifier(n_neighbors=13)
knn.fit(train_x,train_y)

print(knn.score(train_x,train_y))
print(knn.score(test_x,test_y))
```

```python
def mygrid(kn,ver=True):
    
    k=[]
    train_acc=[]
    test_acc=[]
    for i in range(1,kn,2):
        knni=KNeighborsClassifier(n_neighbors=i)
        knni.fit(train_x,train_y)
        k.append(i)
        ktt=knni.score(train_x,train_y)
        kts=knni.score(test_x,test_y)
        train_acc.append(ktt)
        train_acc.append(kts)
        if ver:
            print(f"for k= {i} ; train score {round(ktt,3)} and test score {round(kts,3)}")
return k, train_acc, test_acc
```
